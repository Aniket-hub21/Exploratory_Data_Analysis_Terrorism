```python
# Importing libraries and dependencies

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Reading Dataset

data = pd.read_csv("G:\Internship\globalterrorismdb_0718dist.csv", encoding='latin-1', low_memory=False)
```

```python
# Displaying First 5 Rows of Dataset

data.head()
```

```python
# Checking the Dimensions of Dataset

data.shape
```

```python
# Displaying Column Names

data.columns
```

```python
# Checking Null Values 

data.isnull().sum()
```

```python
# Creating a separate Dataframe to clean the Dataset

df = data.copy()
```

```python
df.head()
```

```python
df.dropna(axis=1, how='all')
```

```python
# Renaming Columns for better understanding

df.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','provstate':'State','region_txt':'Region','attacktype1_txt':'AttackType','target1':'Target','nkill':'Killed','nwound':'Wounded','summary':'Summary','gname':'Group','targtype1_txt':'Target_type','weaptype1_txt':'Weapon_type','motive':'Motive'},inplace=True)
```

```python
# Keeping only necessary Data 

df=df[['Year','Month','Day','Country','State','Region','city','latitude','longitude','AttackType','Killed','Wounded','Target','Summary','Group','Target_type','Weapon_type','Motive']]
```

```python
# Filling NA Values with "0"

df['Wounded']=df['Wounded'].fillna(0).astype(int)
df['Killed']=df['Killed'].fillna(0).astype(int)
```

```python
# Calculating Casualities

df['Casualities'] = df['Killed'] + df['Wounded']
```

```python
df.head()
```

```python
df.isnull().sum()
```

# Data Analysis

Following analysis to be done :

1. Number of Terrorist Activities by Year
2. Attack Types by Terrorist
3. Popular Targets of Terrorist
4. Number of Terrorist activities by Region
5. Terrorist Activities by Region each year
6. Top Countries Affected
7. Number of Attacks and Number of People Killed
8. Terrorist Organisation with Highest Terror Attacks

# Number of Terrorist Activities by Year

```python
plt.subplots(figsize=(15,8))
sns.countplot('Year',data=df,palette='RdYlGn_r',edgecolor=sns.color_palette("husl", 10))
plt.xticks(rotation=90)
plt.title('Number of Terrorist Activities by Year',fontsize=20)
plt.xlabel('Year',fontsize=15)
plt.ylabel('Count',fontsize=15)
plt.show()
```

We can see that between 1998 and 2004 there are low terrorist attacks but after that there is regular increament in terrorist activities with highest activities occurred in 2014.

# Attack Types by Terrorist

```python
plt.subplots(figsize=(15,6))
sns.countplot('AttackType',data=df,palette='cool',order=df['AttackType'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Attack types by Terrorist',fontsize=20)
plt.xlabel('Attack Type',fontsize=15)
plt.ylabel('Count',fontsize=15)
plt.show()
```

Bombing_explosion is the most used attack type by the terrorist followed by Armed Assault and Assasination and least attack used are Hostage Taking and Hijacking.

# Popular Targets of Terrorists

```python
plt.subplots(figsize=(18,6))
sns.countplot('Target_type',data=df,palette='rocket',order=df['Target_type'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Popular Targets of Terrorists',fontsize=20)
plt.xlabel('Target',fontsize=15)
plt.ylabel('Count',fontsize=15)
plt.show()
```

Private Citizens and Property is the most popular target for terrorists.

# Number of Terrorist activities by Region

```python
plt.subplots(figsize=(18,8))
sns.countplot('Region',data=df,palette='deep',edgecolor=sns.color_palette("dark", 10),order=df['Region'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Number of Terrorist activities by Region',fontsize=20)
plt.xlabel('Region',fontsize=15)
plt.ylabel('Count',fontsize=15)
plt.show()
```

We can see Middle East & North Africa is the most affected region followed by South Asia and South America.

# Terrorist Activities by Region each year

```python
pd.crosstab(df.Year, df.Region).plot(kind='area',figsize=(15,6))
plt.title('Terrorist Activities by Region each year', fontsize=18)
plt.xlabel('Year', fontsize= 15)
plt.ylabel('Number of Attacks', fontsize= 15)
plt.show()
```

In recent years Middle East and South Asia are the victim of the most terrorirst attacks but during the 70's and 80's Central America and Western Europe were the main victim of terrorism. Also the number of terrorist attacks has been increased sharply after 2010 largely because of the increased terrorist activities in middle east for the regions' unsettled political environment.

# Top Countries Affected

```python
plt.subplots(figsize=(20,8))
sns.barplot(df['Country'].value_counts()[:15].index,df['Country'].value_counts()[:15].values,palette='cool')
plt.title('Top Countries Affected',fontsize=20)
plt.xlabel('Countries',fontsize=15)
plt.ylabel('Count',fontsize=15)
plt.xticks(rotation= 90)
plt.show()
```

Iraq is the most affected country followed by Pakistan and Afganistan.

# Number of Attacks and Number of People Killed

```python
count_terror = df['Country'].value_counts()[:15].to_frame()
count_terror.columns=['Attacks']
count_kill=df.groupby('Country')['Killed'].sum().to_frame()
count_terror.merge(count_kill,left_index=True,right_index=True,how='left').plot.bar(width=0.9)
fig=plt.gcf()
fig.set_size_inches(20,8)
plt.title('Number of Attacks and Number of People Killed',fontsize=20)
plt.xlabel('Countries',fontsize=15)
plt.ylabel('Count',fontsize=15)
plt.show()
```

From the above barplot we can get a simple overview about how deadly terrorists attacks in Iraq that has taken thousands of lives.

# Terrorist Organisation with Highest Terror Attacks

```python
sns.barplot(df['Group'].value_counts()[1:20].values,df['Group'].value_counts()[1:20].index,palette=("copper"))
plt.xticks(rotation=90)
fig=plt.gcf()
fig.set_size_inches(10,8)
plt.title('Terrorist Organisation with Highest Terror Attacks')
plt.show()
```

Taliban seems to be the most active terror group.
